       ## Let's create a prediction model, trying to predic a players value and/or overall ranking based on Age, Reactions, Composure and Potential.  

```python
#Separating machine learning data sets with columns
df_ml = df[['Age' ,'Reactions' , 'Composure' , 'Potential', 'Overall', 'Value']]
```

```python
df_ml.head()
```

    ### Data Cleaning

    ## remove non-numeric values 

```python
df_ml['Value'] = df_ml['Value'].str.replace(r"[a-zA-Z]",'')
```

```python
df_ml['Value'] = df_ml['Value'].str.replace(r"€", '')
```

```python
df_ml.head()
```

    ## Deal with Missing Values

```python
df_ml.info()
```

```python
#Fill Missing Values with Mean per columns 
import pandas as pd
import numpy as np

from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)

X = pd.DataFrame(df_ml)
df_ml = DataFrameImputer().fit_transform(X)
```

```python
df_ml.info()
```

```python
df_ml.describe().transpose()
```

```python
df_ml.columns
```

    ## Splitting Train and Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
features = df_ml[['Age', 'Reactions', 'Composure', 'Potential', 'Overall']]
label = df_ml[['Value']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=101)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

    ## Using a pipeline to fit transfor training data

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")), #it will take care of missing values - which we don't have here anymore
        ('std_scaler', StandardScaler()),
    ])

X_train_tr = num_pipeline.fit_transform(X_train)
```

```python
X_train_tr
```

    ## transforming test data set

```python
X_test_tr = num_pipeline.transform(X_test)
```

```python
X_test_tr
```

    ### Linear Regression Model

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train_tr, y_train)
```

```python
#Make predictions
df_predictions = lin_reg.predict(X_test_tr)
 
#Evaluate Model - rmse
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_test, df_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
#Evaluate Model - Mean Absolute Error
from sklearn.metrics import mean_absolute_error
lin_mae = mean_absolute_error(y_test, df_predictions)
lin_mae
```

```python
#Fine Tune Model with Cross Validation
#Fine Tune Cross Validation
from sklearn.model_selection import cross_val_score
lin_scores = cross_val_score(lin_reg, X_train_tr, y_train, scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
display_scores(lin_rmse_scores)
```
